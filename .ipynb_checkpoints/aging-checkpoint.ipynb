```python
import pandas as pd
from datetime import datetime
import pytz
df = pd.read_csv("Pendency_automation_report_ageing_greater_than_12.csv")
df
```

```python
df['fact_updated_at'] = pd.to_datetime(df['fact_updated_at'])

time_zone = pytz.timezone('Asia/Kolkata')
current_time = datetime.now(time_zone)

df['hour_diff'] = (current_time - df['fact_updated_at']) / pd.Timedelta(hours=1)

df['hour_diff'] = df['hour_diff'].round().astype(int)
df['hour_diff']
```

```python
def categorize_time(hour_diff):
    if hour_diff > 48:
        return '> 48 hours'
    elif hour_diff > 24:
        return '24-48 hours'
    elif hour_diff > 11:
        return '12-24 hours'
    else:
        return '< 12 hours'
    
df["time_category"] = df['hour_diff'].apply(categorize_time)

data = df['time_category'].value_counts()

print(data)


# data1 = df[['hour_diff', 'time_category']]

# print(data1)
```
